```python
import numpy as np
random_int=pd.Series(np.random.randint(1, 5,size=365), name='random_int')
random_int.head()
```

```python
nama_hari0=['Monday','Tuesday','Wednesday','Thursday','Friday']
nama_hari1=np.random.choice(nama_hari0, size=365)
nama_hari=pd.Series(nama_hari1, name='nama_hari')
nama_hari
```

```python
df3=pd.concat([random_int, nama_hari], axis=1)
df3.head()
```

```python
df4=df3.groupby(['nama_hari'])['random_int'].count()
df4.head()
```
